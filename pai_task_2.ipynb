```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_file = pd.read_csv("train.csv")
test_file = pd.read_csv('test.csv')
```

```python
train_file
```

```python
test_file
```

```python
train_file.drop(['PassengerId', 'Name'], axis=1, inplace=True)
test_file.drop(['PassengerId', 'Name'], axis=1, inplace=True)

```

```python
train_file
```

```python
train_file.isnull().sum()
```

```python
test_file.isnull().sum()
```

```python
train_file.fillna({
    'HomePlanet': train_file['HomePlanet'].mode()[0],
    'CryoSleep': train_file['CryoSleep'].mode()[0],
    'Cabin': train_file['Cabin'].mode()[0],
    'Destination': train_file['Destination'].mode()[0],
    'VIP': train_file['VIP'].mode()[0],
    'Age': train_file['Age'].median(),
    'RoomService': train_file['RoomService'].median(),
    'FoodCourt': train_file['FoodCourt'].median(),
    'ShoppingMall': train_file['ShoppingMall'].median(),
    'Spa': train_file['Spa'].median(),
    'VRDeck': train_file['VRDeck'].median()
}, inplace=True)
```

```python
test_file.fillna({
    'HomePlanet': test_file['HomePlanet'].mode()[0],
    'CryoSleep': test_file['CryoSleep'].mode()[0],
    'Cabin': test_file['Cabin'].mode()[0],
    'Destination': test_file['Destination'].mode()[0],
    'VIP': test_file['VIP'].mode()[0],
    'Age': test_file['Age'].median(),
    'RoomService': test_file['RoomService'].median(),
    'FoodCourt': test_file['FoodCourt'].median(),
    'ShoppingMall': test_file['ShoppingMall'].median(),
    'Spa': test_file['Spa'].median(),
    'VRDeck': test_file['VRDeck'].median()
}, inplace=True)
```

```python
train_file.isnull().sum()
```

```python
test_file.isnull().sum()
```

```python
train_file.dtypes
```

```python
categorical_cols = ['HomePlanet', 'Destination']
train_data = pd.get_dummies(train_file, columns=categorical_cols, drop_first=True)
```

```python
categorical_cols = ['HomePlanet', 'Destination']
test_data = pd.get_dummies(test_file, columns=categorical_cols, drop_first=True)
```

```python
train_data.dtypes
```

```python
test_data.dtypes
```

```python
train_data
```

```python
train_data.drop(['Cabin'], axis=1, inplace=True)
```

```python
test_data.drop(['Cabin'], axis=1, inplace=True)
```

```python
test_data
```

```python
test_data
```

```python
train_data.dtypes
```

```python
for col in train_data.columns:
    if train_data[col].dtype == 'bool':
        train_data[col] = train_data[col].astype(int)
    elif train_data[col].dtype == 'float64':
        train_data[col] = train_data[col].astype(int)
```

```python
for col in test_data.columns:
    if test_data[col].dtype == 'bool':
        test_data[col] = test_data[col].astype(int)
    elif test_data[col].dtype == 'float64':
        test_data[col] = test_data[col].astype(int)
```

```python
x = train_data.drop(['Transported'], axis=1)
y = train_data['Transported'].astype(int)
```

```python
RFC = RandomForestClassifier(random_state=42)
RFC.fit(x, y)
```

```python
predictions = RFC.predict(test_data)
dump = pd.read_csv('test.csv')
submission = pd.DataFrame({'PassengerId':dump['PassengerId'],'Transported':predictions})
submission.to_csv('Submission.csv',index = False )
```
